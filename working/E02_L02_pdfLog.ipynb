```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
a = 0.1
b = 10
N = 100000
x = np.random.uniform(a,b,N)
```

```python
nbins=20
plt.hist(x,nbins)
plt.xlabel('$x$')
plt.show()
```

```python
y = np.log10(x)
```

```python
nbins=20
plt.hist(y,nbins,density = True)
plt.xlabel('$y$')
plt.show()
```

Theoretical derivation of the distribution for y:

$$ p(x) = \frac{1}{b-a}$$ if $x \in [a,b]$

$$ y = \log_{10}x $$

$$ dy = \frac{1}{x \ln(10)} dx $$

$$ p(y) = \frac{dx}{dy} p(x) =  10^y \ln(10) \frac{1}{b-a} $$


```python
def pdf_y(y): return 10**y * np.log(10) / (b-a)

y_grid = np.linspace(-1,1,100)

nbins=20
plt.hist(y, nbins, density = True, label='Data')
plt.plot(y_grid, pdf_y(y_grid), label='Pdf')
plt.xlabel('$y$')
plt.legend()
plt.show()
```

```python
np.log10(np.mean(x)), np.mean(y)
```

```python
np.log10(np.median(x)), np.median(y)
```

The mean changes because the mean of the log is different from the log of the mean. Instead, the median doesn't change because the transformation is monotone, so the ordering doesn't change (and the median depends on the ordering).
