```python
import numpy as np
import pylab as plt
import scipy
```

```python
deaths=np.arange(0,5)
groups=np.array([109,65,22,3,1])
freq_groups=groups/np.sum(groups)
```

```python
plt.scatter(deaths,freq_groups)
plt.xticks(deaths)
plt.xlabel('Deaths')
plt.ylabel('Frequence of groups')
plt.title('Horse kicks')
plt.show()
```

```python
# As mean I use the average weighted over the frequency of groups
mu = np.average(deaths, weights=freq_groups)
poisson = scipy.stats.poisson(mu)
print(mu)
```

```python
plt.plot(deaths, poisson.pmf(deaths), color='orange')
plt.scatter(deaths,freq_groups)
plt.xticks(deaths)
plt.xlabel('Deaths')
plt.ylabel('Frequence of groups')
plt.title('Horse kicks')
plt.show()
```
