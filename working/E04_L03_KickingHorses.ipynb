```python
import numpy as np
import pylab as plt
import scipy.stats

plt.style.use('seaborn-v0_8-deep')
plt.rc('text', usetex=True)  
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
deaths = np.array([0,1,2,3,4])
groups = np.array([109,65,22,3,1])

tot = np.sum(groups)
print(tot)
```

```python
freq = groups/tot
```

```python
plt.scatter(deaths, freq)
plt.xticks(deaths)
plt.xlabel('Deaths')
plt.ylabel('Frequence')
plt.title('Horse kicks')
plt.show()
```

To try and fit by hand, I use the poissonian distribution with the sample mean as an estimator of mu.

```python
mu = np.sum(deaths*groups)/np.sum(groups)
print(mu)

poisson = scipy.stats.poisson(mu)
```

```python
plt.plot(deaths, poisson.pmf(deaths), color='C1', label='pdf')
plt.scatter(deaths, freq, label='Data')
plt.xticks(deaths)
plt.xlabel('Deaths')
plt.ylabel('Frequency')
plt.title('Horse kicks')
plt.legend()

plt.show()
```

Compute the fit and compare the values of mu.

```python
from scipy.optimize import curve_fit

def poisson_pmf(x, mu):
    return scipy.stats.poisson.pmf(x, mu)

params, cov = curve_fit(poisson_pmf, deaths, freq)
fitted_mu = params[0]
print('Fitted mu =', fitted_mu)
print('Sample mean =', mu)
```
